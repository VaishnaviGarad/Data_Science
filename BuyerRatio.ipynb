Chi Square test of independence for categorical data

```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
import statsmodels.api as sm
from scipy.stats import chi2_contingency
from scipy.stats import norm
```

```python
data = pd.read_csv('BuyerRatio (1).csv')
data.head()
```

```python
data = data.drop('Observed Values', axis=1)
data
```

```python
stats, p, dof, expfreq = stats.chi2_contingency(data)
print( stats, '\n', p, '\n', dof, '\n', expfreq)
```

```python
alpha = 0.05 #given
print('Chi2 test value is ', p)
if p < alpha:
  print('Reject Null Hypothesis') #--given
else:
  print('Accept Null Hypothesis')
```

# **As we failed to reject null hypothesis, male and female buyer rations are similar across regions.**
